```python
import numpy as np
import scipy.interpolate as sinterp
import matplotlib.pyplot as plt
```

храним двумерный массив в виде пар точек (x, y)
тогда строки - точки
столбцы - вектора значений

```python
x = np.array([1,2,5,8])
y = np.array([2,3,4,3.2])
```

```python
points = np.array([x,y]).T
points
```

хотим удалить строку с определенным значением x

```python
a = points[points[:,0].argsort()]
a
```

```python
spl = sinterp.make_interp_spline(a[:,0], a[:,1], k=1)
```

```python
spl
```

```python
xx = np.linspace(0, 20, 100)
plt.plot(xx, spl(xx))
plt.plot(x,y)
```

```python
p = sinterp.PPoly.from_spline(spl)
```

```python
p.roots()
```

```python
f = sinterp.interp1d(a[:,0], a[:,1],fill_value="extrapolate")
```

```python
xx = np.linspace(0, 20, 100)
plt.plot(xx, f(xx))
plt.plot(x,y)
```
